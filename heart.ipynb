```python
X = data.drop('target', axis = 1)
y = data['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
print(X_train.shape)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
y_predict
```

```python
y_test
```

```python
model.score(X_test,y_test)
```

```python
from xgboost import XGBClassifier
```

```python
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb_model.predict(X_test)
y_pred_xgb
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cr2 = classification_report(y_test, y_pred_xgb)
print(cr2)
```

```python
cm2 = confusion_matrix(y_test, y_pred_xgb)
cm2
```

```python
cm2 = np.array([[106,  0],
       [ 0,  99]])

cm2_percent = cm2 / cm2.sum() * 100

labels = np.array([["{0}\n{1:.1f}%".format(value, percent)
                   for value, percent in zip(row, row_percent)]
                  for row, row_percent in zip(cm2, cm2_percent)])

# Create the heatmap
plt.figure(figsize=(10,8))
sns.heatmap(cm2_percent, annot=labels, fmt='', cmap='Blues', cbar_kws={'label': 'Percentage'})

# formatting
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('XGB Confusion Matrix (in %)')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
rf_pred = rf.predict(X_test)
rf_pred
```

```python
cr3 = classification_report(y_test, rf_pred)
print(cr3)
```

```python
cm3 = confusion_matrix(y_test, rf_pred)
cm3
```

```python
cm3 = np.array([[106,  0],
       [ 0,  99]])

cm3_percent = cm3 / cm3.sum() * 100

labels = np.array([["{0}\n{1:.1f}%".format(value, percent)
                   for value, percent in zip(row, row_percent)]
                  for row, row_percent in zip(cm3, cm3_percent)])

# Create the heatmap
plt.figure(figsize=(10,8))
sns.heatmap(cm3_percent, annot=labels, fmt='', cmap='coolwarm', cbar_kws={'label': 'Percentage'})

# formatting
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Random Forest Confusion Matrix (in %)')
plt.show()
```

```python
cm4 = confusion_matrix(y_test, y_pred_dt)
cm4
```
